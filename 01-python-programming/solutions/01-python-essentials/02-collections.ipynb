## Dictionaries

````python
course = {
    'name': 'Python for Young Data Scientists',
    'venue': 'The Connection',
    'cost': 999
}

print(course['name'])
print(course['venue'])
print(course['cost'])
````

````python
course['instructor']
````

````python
course['instructor'] = 'Kan Ouivirach'

print(course)
````

````python
course['instructor']
````

````python
courses = [
    {
        'name': 'Python for Young Data Scientists',
        'venue': 'The Connection 1',
        'cost': 999
    },
    {
        'name': 'R for Young Data Scientists',
        'venue': 'The Connection 2',
        'cost': 999
    },
]

print(courses[1]['name'])
print(courses[1]['venue'])
print(courses[1]['cost'])
````

## Unpacking

````python
item_list = [1, 2, 3]
a = item_list[0]
b = item_list[1]
c = item_list[2]

print(a, b, c)
````

````python
a, b, c = [1, 2, 3]
print(a, b, c)
````

````python
a, b, c = (4, 5, 6)
print(a, b, c)
````

````python
d = {'a': 10, 'b': 20, 'c': 30}
a, b, c = {'a': 10, 'b': 20, 'c': 30}
print(a, b, c)
````

````python
d[c]
````

````python
a, _, c = (4, 5, 6)
a + c
````

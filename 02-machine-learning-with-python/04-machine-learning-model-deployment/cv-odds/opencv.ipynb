```python
import cv2
```

```python
image = cv2.imread('cherprang087.jpg')
```

```python
image.shape
```

```python
#cv2.imshow('Image', image)
#cv2.waitKey(0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(rgb_image)
```

```python
rgb_image[100, 200]
```

```python
roi = rgb_image[50:350, 150:450]
plt.figure(figsize=(3, 3))
plt.imshow(roi)
```

```python
new_image = rgb_image.copy()
cv2.rectangle(new_image, (150, 50), (450, 350), (255, 0, 0), 5)
plt.figure(figsize=(3, 3))
plt.imshow(new_image)
```

```python
image_with_text = new_image.copy()
cv2.putText(
    image_with_text, 
    'Hello', 
    (200, 400), 
    cv2.FONT_HERSHEY_COMPLEX, 
    3, 
    (0, 255, 0),
    5
)
plt.figure(figsize=(3, 3))
plt.imshow(image_with_text)
```
